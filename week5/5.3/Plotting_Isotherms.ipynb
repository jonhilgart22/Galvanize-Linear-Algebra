```python
def function(x,y):
    return x**2-4*x-y**2
```

```python
question_one_data = [function(x,y) for x  in range(-100,100) for y in range(-50,50)]
```

```python
import seaborn as sns
%pylab inline
```

```python
sns.distplot(question_one_data)
```
